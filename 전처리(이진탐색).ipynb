```python
from datasets import load_dataset
import pandas as pd

# 데이터셋 로드
train_dataset = load_dataset("jeanlee/kmhas_korean_hate_speech", split="train")
validation_dataset = load_dataset("jeanlee/kmhas_korean_hate_speech", split="validation")
test_dataset = load_dataset("jeanlee/kmhas_korean_hate_speech", split="test")

# 데이터셋을 데이터프레임으로 변환
train_df = pd.DataFrame(train_dataset)
validation_df = pd.DataFrame(validation_dataset)
test_df = pd.DataFrame(test_dataset)

# 데이터프레임 합치기
combined_df = pd.concat([train_df, validation_df, test_df], ignore_index=True)

# 합친 데이터프레임을 CSV 파일로 저장
combined_df.to_csv('combined_korean_hate_speech.csv', index=False)
```

```python
import pandas as pd
import ast

# CSV 파일 불러오기
combined_df = pd.read_csv('combined_korean_hate_speech.csv')

# 라벨이 리스트 형식으로 저장되어 있으므로 이를 리스트로 변환
combined_df['label'] = combined_df['label'].apply(ast.literal_eval)

# 'label'이 0에서 7 사이의 숫자를 포함하는 데이터만 선택
def is_hate(labels):
    return all(label in range(8) for label in labels)

hate_df = combined_df[combined_df['label'].apply(is_hate)]

# 'label'을 모두 1으로 설정
hate_df['label'] = 1

# 'label'과 'text' 열만 선택하고 순서를 'label', 'text'로 설정
hate_df = hate_df[['label', 'text']]

# 결과 출력 (처음 몇 줄만)
print(hate_df.head())
print(hate_df.shape)
```

```python
import pandas as pd
import ast

# CSV 파일 불러오기
combined_df = pd.read_csv('combined_korean_hate_speech.csv')

# 라벨이 리스트 형식으로 저장되어 있으므로 이를 리스트로 변환
combined_df['label'] = combined_df['label'].apply(ast.literal_eval)

# 'label'이 8만 포함하는 데이터만 선택
def is_nonhate(labels):
    return all(label == 8 for label in labels)

nonhate_df = combined_df[combined_df['label'].apply(is_nonhate)]

# 'label'을 모두 0으로 설정
nonhate_df['label'] = 0

# 'label'과 'text' 열만 선택하고 순서를 'label', 'text'로 설정
nonhate_df = nonhate_df[['label', 'text']]

# 결과 출력 (처음 몇 줄만)
print(nonhate_df.head())
print(nonhate_df.shape)
```

```python
import pandas as pd
from transformers import BertTokenizer
from keras.preprocessing.sequence import pad_sequences
import torch

# BERT 토크나이저 불러오기
tokenizer = BertTokenizer.from_pretrained('bert-base-multilingual-cased', do_lower_case=False)

MAX_LEN = 128

# test_data.csv 파일 불러오기 (이미 생성된 파일)
test_data = pd.read_csv('test_data.csv')

# 데이터 확인 (예시로 처음 몇 줄 출력)
print(test_data.head())

# 토큰화 및 텐서 변환 함수 정의
def data_to_tensor(sentences, labels):
    # 문장을 토큰화하여 토큰 ID로 변환
    tokenized_texts = [tokenizer.tokenize(str(sent)) for sent in sentences]  # str으로 변환하여 안전성 확보
    input_ids = [tokenizer.convert_tokens_to_ids(x) for x in tokenized_texts]

    # 패딩을 추가하여 길이를 맞춤
    input_ids = pad_sequences(input_ids, maxlen=MAX_LEN, dtype="long", truncating="post", padding="post")

    # 어텐션 마스크 생성
    attention_masks = [[float(i > 0) for i in seq] for seq in input_ids]

    # 텐서로 변환
    tensor_inputs = torch.tensor(input_ids)
    tensor_labels = torch.tensor(labels)
    tensor_masks = torch.tensor(attention_masks)

    return tensor_inputs, tensor_labels, tensor_masks

# 테스트 데이터셋을 토큰화하여 텐서로 변환
test_inputs, test_labels, test_masks = data_to_tensor(test_data['text'], test_data['label'])

# 결과 출력 (예시로 처음 몇 줄 출력)
print("Test Input IDs:")
print(test_inputs[:3])
print("Test Labels:")
print(test_labels[:3])
print("Test Attention Masks:")
print(test_masks[:3])
```
